```python
import pandas as pd
import urllib, json
import urllib2  # the lib" that handles the url stuff
import re
import numpy as np
```

```python
def dataframe(state_code, YYYY):
    variables = 'http://api.census.gov/data/2015/acs5/variables.json'
    json_variables = urllib.urlopen(variables)
    variables_data = json.loads(json_variables.read())
    variables_list =[]
    label = {}
    for key in variables_data['variables'].keys():
        if key.startswith('B20001') and key.endswith('E'):
            variables_list.append(key)
            label[key] = variables_data['variables'][key]['label']
    base_url = 'http://api.census.gov/data/' + str(YYYY) + '/acs5?get='
    url = base_url + ','.join(variables_list) + '&for=county+subdivision:*&in=state:' + str(state_code)
    json_url = urllib.urlopen(url)
    url_data = json.loads(json_url.read())
    title = []
    for code in url_data[0]:
        if code in label:
            title.append(label[code])
        else:
            title.append(code)
    county_dic={}
    county_url = 'http://www2.census.gov/geo/docs/reference/codes/files/st25_ma_cousub.txt'
    county = urllib.urlopen(county_url)
    county = county.readlines()
    for line in county:
        c = line.split(',')
        county_dic[c[4]]= c[5]
    income = pd.DataFrame(url_data[2:], columns = title)
    for key in county_dic.keys():
        income['county subdivision'].replace(key, county_dic[key], inplace = True)
    income.set_index('county subdivision', inplace = True)
    income = income.astype(int)
    income = income.reindex_axis(sorted(income.columns), axis =1)

    return income
    
```
